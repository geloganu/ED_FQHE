# Energy Spectrum

### $\nu = 5/2$ No Disorder

```python
filename = "./data/N14_Spectrum.npy"
ene_spectrum14 = np.load(filename)

filename = "./data/N12_Spectrum.npy"
ene_spectrum12 = np.load(filename)

L12 = ene_spectrum12[:,0]
ene12 = ene_spectrum12[:,1]

L14 = ene_spectrum14[:,0]
ene14 = ene_spectrum14[:,1]
```

```python

from matplotlib.ticker import AutoMinorLocator

fig, ax = plt.subplots(1,2,figsize=(11, 5))

ax[0].plot(L12, ene12-ene12.min(), ls="none", marker="_", ms="15", mew="1.5")
ax[1].plot(L14, ene14-ene14.min(),ls="none", marker="_", ms="15", mew="1.5")

ax[0].set_xlabel("L")
ax[1].set_xlabel("L")
ax[0].set_ylabel("Energy")
ax[1].set_ylabel("Energy")

ax[0].grid(which='major', color='black', linestyle='--', linewidth=0.5,alpha=0.2)
ax[0].grid(which='minor', color='black', linestyle='--', linewidth=0.5,alpha=0.2)
ax[1].grid(which='major', color='black', linestyle='--', linewidth=0.5,alpha=0.2)
ax[1].grid(which='minor', color='black', linestyle='--', linewidth=0.5,alpha=0.2)

ax[0].xaxis.set_minor_locator(AutoMinorLocator(2))
ax[0].yaxis.set_minor_locator(AutoMinorLocator(2))
ax[1].xaxis.set_minor_locator(AutoMinorLocator(2))
ax[1].yaxis.set_minor_locator(AutoMinorLocator(2))

ax[0].text(-1.6, 0.115, 'A', fontsize=15)
ax[1].text(-1.7, 0.093, 'B', fontsize=15)


plt.show()
```
